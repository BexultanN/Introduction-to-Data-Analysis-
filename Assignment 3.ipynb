3.  Write a python program to read the last n lines of file, bear.txt. \\
n=2, 3

```python
n = int(input("Enter n lines:-   "))
f = open("bear.txt", "r")
for line in (f.readlines() [-n:]):
    print(line, end="")
    f.close()
```

7. Write a program to calculate the frequency of each word in the text in your file bear.txt

```python
f =open("bear.txt", "r")
wrd = input("Enter the word for calculate frequency:    ")
s=f.read()
lst=s.split()
count = 0
for i in lst:
    if(i==wrd):
        count+=1
print("FREQUENCY OF THE WORD {} is {} TIMES".format(wrd, count))
```
